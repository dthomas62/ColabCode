```python
# Run this cell to authenticate with Google Cloud to access the GCS bucket
from google.colab import auth
auth.authenticate_user()

import pandas as pd
```

```python
import os
file_path = os.environ.get('RADAR_REQUEST_PATH', 'gs://converged-brandradar/lookups/RadarRequest.xlsx')

# 1. Job details
# Read without header so we can access absolute indices easily
try:
    df_job = pd.read_excel(file_path, sheet_name='Job details', header=None)
except ValueError:
    # Fallback just in case sheet is strictly lowercase
    df_job = pd.read_excel(file_path, sheet_name='job details', header=None)

# iloc[row, col] -> Column B is index 1.
# Row 3 -> index 2, Row 9 -> index 8, etc.
job_name = df_job.iloc[2, 1]
brand_radar = df_job.iloc[8, 1]
social_radar = df_job.iloc[9, 1]
culture_connects = df_job.iloc[10, 1]
culture_converts = df_job.iloc[11, 1]

print("--- Job Details ---")
print(f"Job Name: {job_name}")
print(f"BrandRadar: {brand_radar}")
print(f"SocialRadar: {social_radar}")
print(f"Culture Connects: {culture_connects}")
print(f"CultureConverts: {culture_converts}")
```

```python
# 2. Brand details (Columns B: Brand, C: Category, D: Market)
try:
    dfbrands = pd.read_excel(file_path, sheet_name='Brand details', usecols="B:D")
except ValueError:
    dfbrands = pd.read_excel(file_path, sheet_name='brand details', usecols="B:D")

dfbrands = dfbrands.dropna(how='all') # Clean up any empty/blank rows

print("\n--- Brand Details (dfbrands) ---")
display(dfbrands)
```

```python
# 3. Target Audiences (Column B: Market, Column C: Audience)
try:
    dfaudience = pd.read_excel(file_path, sheet_name='Target Audiences', usecols="B:C")
except ValueError:
    dfaudience = pd.read_excel(file_path, sheet_name='target audiences', usecols="B:C")

dfaudience = dfaudience.dropna(how='all') # Clean up any empty/blank rows

print("\n--- Target Audiences (dfaudience) ---")
display(dfaudience)
```
